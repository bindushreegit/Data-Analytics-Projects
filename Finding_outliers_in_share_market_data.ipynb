### Assignment_6: Finding outliers in share market data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Specify the range of rows you want to read
start_row = 4 
end_row = 11   

# Read the specific range of rows from the Excel file
data = pd.read_excel('C:\\Users\\Bindu\\Documents\\Arcada\\Visual_Analytics\\Assainments\\Ass6\\Share Price data for Assignment regarding Manipulation.xlsx', skiprows=start_row-1, nrows=end_row-start_row+1)

# Drop columns with all NaN values
data.dropna(axis=1, how='all', inplace=True)

# Melt the DataFrame to convert it to a long format suitable for plotting
data_melted = data.melt(id_vars='Date', var_name='Time', value_name='SharePrice')

print(data_melted)

elizabeth_trading_days = data_melted[(data_melted['Date'] == '14th April') | (data_melted['Date'] == '18th April')]
#print(elizabeth_trading_days)
```

```python
# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Date', y='SharePrice', data=data_melted)
plt.title('Hourly Share Prices - Box Plot',fontsize=15,fontweight='bold')
plt.xlabel('Date',fontsize=15,fontweight='bold')
plt.ylabel('Share Price',fontsize=15,fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Summary:
The hourly share price data for a week indicates potential irregularities in trading activities on specific days, notably the 14th and 19th of April. A box plot visualization displays outliers on these days, hinting at significant deviations from the typical share prices observed throughout the week. This prompts suspicion of potential manipulation or hacking of the share trading software, particularly during periods of active trading by Elizabeth.

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Time', y='SharePrice', hue='Date', data=elizabeth_trading_days, dodge=True)
plt.title('Hourly Share Prices - Swarm Plot',fontsize=15,fontweight='bold')
plt.xlabel('Time',fontsize=15,fontweight='bold')
plt.ylabel('Share Price',fontsize=15,fontweight='bold')
plt.legend(title='Date', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
