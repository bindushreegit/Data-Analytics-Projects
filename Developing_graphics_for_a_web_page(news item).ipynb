## Assignment #5: Developing graphics for a web page (news item)

```python
import matplotlib.pyplot as plt


x_millenials = [60,33,7]
y_millenials = ['Support','Oppose','Neutral']

x_genz = [40,50, 10]
y_genz = ['Support','Oppose','Neutral']

#x_gen45plus = [33,50,]
#y_gen45plus = ['Suport','Oppose']

# Color palette
colors = ["olive", "orangered", 'whitesmoke']
ex=[0.,0.,0.2]

fig = plt.figure(figsize=(10, 6))
gs = fig.add_gridspec(1, 2, hspace=0., wspace=0.6)
(ax1, ax2) = gs.subplots(sharex='col', sharey='row')
fig.suptitle('Millennials Want Gender Pronouns on Work Emails, but Many in Gen Z Dont \n (Poll of 1,500 U.S. adults by Redfield & Wilton Strategies for Newsweek)',fontsize = 15,fontweight='bold')
ax1.pie(x_millenials, labels=y_millenials,colors=colors,explode=ex,autopct="%0.1f%%",shadow=True,radius=1.,startangle=280,textprops={"fontsize":15})
ax1.set_title('% of Millenials aged 25-44 years',y=1.01,fontweight='bold')
ax2.pie(x_genz, labels=y_genz,colors=colors,explode=ex,autopct="%0.1f%%",shadow=True,radius=1.,startangle=265,textprops={"fontsize":15})
ax2.set_title('% of Generation Z aged 18-24 years',y=1.01,fontweight='bold')
#plt.suptitle('Generation-wise Favorability and Disfavorability')
plt.legend(loc=1,bbox_to_anchor=(1.1, 1.),fancybox=True, shadow=True)
plt.savefig('NewsWeek_Poll.png',bbox_inches='tight',dpi=600)
plt.show()

```
